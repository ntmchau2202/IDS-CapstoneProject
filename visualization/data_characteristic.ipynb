```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load data
real_estate_df = pd.read_csv("../cleaned/realestate/real_estate_cleaned.csv")
restaurant_df = pd.read_csv("../cleaned/restaurant/raw_restaurant_categorized_cleaned.csv")
```

```python
# District vs num of restaurant
restaurant_count_by_district = restaurant_df[["location","No."]].groupby(["location"]).count().reset_index()

districts = restaurant_count_by_district["location"].to_numpy()
count = restaurant_count_by_district["No."].to_numpy()
y_pos = np.arange(len(districts))

plt.barh(y_pos, count, align='center', alpha=0.5)
plt.yticks(y_pos, districts)
plt.xlabel('Restaurants')
plt.title('Number of restaurant in each district')

plt.show()
```

```python
# Category vs num of restaurant
restaurant_count_by_category = restaurant_df[["category","No."]].groupby(["category"]).count().reset_index()

categorys = restaurant_count_by_category["category"].to_numpy()
count = restaurant_count_by_category["No."].to_numpy()
y_pos = np.arange(len(categorys))

plt.barh(y_pos, count, align='center', alpha=0.5)
plt.yticks(y_pos, categorys)
plt.xlabel('Restaurants')
plt.title('Number of restaurant in each category')

plt.show()
```

```python
# Price rank vs num of restaurant
restaurant_count_by_price = restaurant_df[["price","No."]].groupby(["price"]).count().reset_index()

price_rank = restaurant_count_by_price["price"].to_numpy()
count = restaurant_count_by_price["No."].to_numpy()
y_pos = np.arange(len(price_rank))

plt.barh(y_pos, count, align='center', alpha=0.5)
plt.yticks(y_pos, price_rank)
plt.xlabel('Restaurants')
plt.title('Number of restaurant in each price rank')

plt.show()
```

```python
# District vs avarage rating of restaurant
restaurant_mean_rating_by_district = restaurant_df[["location","rating"]].groupby(["location"]).mean().reset_index()

districts = restaurant_mean_rating_by_district["location"].to_numpy()
mean = restaurant_mean_rating_by_district["rating"].to_numpy()
y_pos = np.arange(len(districts))

plt.barh(y_pos, mean, align='center', alpha=0.5)
plt.yticks(y_pos, districts)
plt.xlabel('Rating Point')
plt.title('Average rating of restaurant in each district')

plt.show()
```

```python
# Category vs avarage rating of restaurant
restaurant_mean_rating_by_category = restaurant_df[["category","rating"]].groupby(["category"]).mean().reset_index()

categories = restaurant_mean_rating_by_category["category"].to_numpy()
mean = restaurant_mean_rating_by_category["rating"].to_numpy()
y_pos = np.arange(len(categories))

plt.barh(y_pos, mean, align='center', alpha=0.5)
plt.yticks(y_pos, categories)
plt.xlabel('Rating Point')
plt.title('Average rating of restaurant in each category')

plt.show()
```

```python
cuisine_types = restaurant_df.drop(columns=["Unnamed: 0","No.","name","rating","location","category","startrange","endrange","price","pricerank"]).columns.to_numpy()
count_by_cuisine = [len(restaurant_df[restaurant_df[cuisine].notnull()]) for cuisine in cuisine_types]

y_pos = np.arange(len(cuisine_types))

plt.barh(y_pos, count_by_cuisine, align='center', alpha=0.5)
plt.yticks(y_pos, cuisine_types)
plt.xlabel('Number of restaurants')
plt.title('Restaurant distribution over cuisine types')

plt.show()
```
